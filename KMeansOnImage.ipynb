```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import pandas as pd
from PIL import Image
from numpy import asarray
```

```python
image=Image.open('dogs.jpeg')
```

```python
image_arr=asarray(image)
```

```python
image_arr.ndim
```

```python
print(len(image_arr))
print(len(image_arr[0]))
print(len(image_arr[0][0]))
```

```python
image_arr.size
```

```python
dataset=[]
for i in range(len(image_arr)):
  for j in range(len(image_arr[i])):
    dataset.append([i,j,sum(image_arr[i][j])])
```

```python
km=KMeans(n_clusters=2)
km.fit(dataset)
```

```python
km.cluster_centers_
```
