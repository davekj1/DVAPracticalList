```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('salary_data.csv')
dataset=[]
yoe=[]
salary=[]
for i in df.values:
  dataset.append([i[0],i[1]])
  yoe.append([i[0]])
  salary.append(i[1])
```

```python
dataset
```

```python
reg=LinearRegression()
reg.fit(yoe,salary)
```

```python
reg.predict([[10]])
```

```python
reg.coef_
```
