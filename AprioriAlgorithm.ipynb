```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
import numpy as np
```

```python
dataset=pd.read_csv("tesco.csv").values
finaldata=[]
for i in dataset:
  arr=[]
  for j in i:
    if j==j:arr.append(j)
  finaldata.append(arr)
finaldata
```

```python
te=TransactionEncoder()
te_df=te.fit(finaldata).transform(finaldata)
te_df
```

```python
df=pd.DataFrame(te_df,columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets=apriori(df,min_support=0.4,use_colnames=True)
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
res=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6)
res
```

```python
res1=res[res['confidence']==1]
```
