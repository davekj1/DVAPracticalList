```python
import matplotlib.pyplot as plt
import pandas as pd
from  sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
df_refined=df.fillna(df.bedrooms.median())
```

```python
df_refined
```

```python
reg=LinearRegression()
reg.fit(df_refined[['area','bedrooms','age']],df_refined[['price']])
```

```python
ans=reg.predict([[4000,5.0,8.5]])[0][0]
```

```python
print(ans)
```

```python
print(reg.coef_)
print(reg.intercept_)
```
