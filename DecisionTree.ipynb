```python
import pandas as pd
```

```python
df=pd.read_csv('Playtennis.csv')
df
```

```python
output=df['Play Tennis']
input=df.drop('Play Tennis',axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
le_outlook=LabelEncoder()
le_temperature=LabelEncoder()
le_humidity=LabelEncoder()
le_wind=LabelEncoder()
input['Outlook_n']=le_outlook.fit_transform(input['Outlook'])
input['Temperature_n']=le_temperature.fit_transform(input['Temperature'])
input['Humidity_n']=le_humidity.fit_transform(input['Humidity'])
input['Wind_n']=le_wind.fit_transform(input['Wind'])
input=input.drop(['Outlook','Temperature','Humidity','Wind'],axis=1)
```

```python
input
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(input,output)
```

```python
model.predict(input)
```
