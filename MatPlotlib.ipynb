```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,10])
y=np.array([2,20])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'d')
plt.show()
```

```python
y=np.array([3,8,5,7])
plt.plot(y,linestyle='dotted')
plt.show()
```

```python
x=np.array([2,3,4,5,6,7])
y=np.array([4,6,8,10,12,14])
plt.xlabel('months')
plt.ylabel('Sales per thousand')
plt.plot(x,y,linestyle='dotted')
plt.title("Graph of sales vs months")
plt.grid()
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([4,5,6,7])
plt.subplot(2,1,1)
plt.plot(x,y)
x=np.array([1,2,3,4])
y=np.array([2,3,4,5])
plt.subplot(2,1,2)
plt.plot(x,y)
plt.suptitle("Combined plot")
```

```python
x=np.array([1,2,3,4,5,6])
y=np.array([5,6,7,8,9,19])
plt.scatter(x,y)
x=np.array([7,8,9,10,11,12])
y=np.array([10,11,12,13,14,15])
plt.scatter(x,y)
```
